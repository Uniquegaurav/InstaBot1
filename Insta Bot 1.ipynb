## Your friend has opened a new Food Blogging handle on Instagram and wants to get famous. He wants to follow a lot of people so that he can get noticed quickly but it is a tedious task so he asks you to help him. As you have just learned automation using Selenium, you decided to help him by creating an Instagram Bot.
You need to create different functions for each task.


```python
# importing required libraries and starting webdriver session and opening
# instagram wepage
from bs4 import BeautifulSoup
import time
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
driver = webdriver.Chrome(executable_path = '/webdrivers/chromedriver')
driver.get('https://www.instagram.com/')
wait = WebDriverWait(driver,10)
```

### 1 Login to your Instagram Handle Submit with sample username and password 

```python
#1
def insta_login() :
    usename =  'SAMPLE USERNAME'
    password = 'SAMPLE PASSWORD'
    box = driver.find_elements_by_class_name('_2hvTZ')
    username_box ,pass_box= box[0] ,box[1]
    username_box.send_keys(usename)
    pass_box.send_keys(password)
    pass_box.submit()
    #Explicit wait until presence of home button and clicking on home for avoiding save info option
    wait = WebDriverWait(driver,15)
    home = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_8-yf5 ')))
    home.click()
    # explicit wait for turn on notification option and clicking on not now
    not_now = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'HoLwm ')))
    not_now.click()
insta_login()
```

### 2  Type for “food” in search bar and print all the names of the Instagram Handles that are displayed in list after typing “food”
Note : Make sure to avoid printing hashtags

```python
#2
def search_food() :
    search_bar  = driver.find_element_by_class_name('x3qfX')
    search_bar.send_keys("food")
    # explicit wait until all insta handle get loaded
    wait = WebDriverWait(driver,20)
    wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,'Ap253')))
    all_results = driver.find_elements_by_class_name('Ap253')
    insta_handle = []
    for i in all_results :
        s = i.text
        if(s[0]!= '#') :
           print(s)
           insta_handle.append(s)
    search_bar.clear()
    driver.refresh()
    for i in insta_handle :
        print(i)
search_food()
```

### 3 Searching and Opening a profile using Open profile of “So Delhi”  


```python
def open_profile(prof) :
    search_bar  = driver.find_element_by_class_name('x3qfX')
    search_bar.clear()
    search_bar.send_keys(prof)
    # explicit wait until all insta handle get loaded
    wait = WebDriverWait(driver,10)
    profile = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'Ap253')))
    profile.click()
driver.refresh()
open_profile('so delhi')
time.sleep(6)
driver.back()
```

### 5 Like/Unlike posts
1 Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.

2 Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python
### 5 1) Liking the top 30 posts of the ‘dilsefoodie'. Print message if #you have already liked it.
def like_posts() :
    open_profile('dilsefoodie')
    time.sleep(2)
    driver.execute_script('window.scrollTo(0,arguments[0]);',0)
    current_height = driver.execute_script('return document.body.scrollHeight;')
    while True :
        driver.execute_script('window.scrollTo(0,arguments[0]);',current_height)
        time.sleep(3)
        new_height = driver.execute_script('return document.body.scrollHeight;')
        new_posts = driver.find_elements_by_class_name('_9AhH0')
        #print(len(new_posts))
        if(len(new_posts)>=30) :
            break;
        if(new_height == current_height) :
            break;
        current_height = new_height
    new_posts = driver.find_elements_by_class_name('_9AhH0')
    driver.execute_script('window.scrollTo(0,arguments[0]);',0)
    new_posts = new_posts[:30]
    #print(len(new_posts))
    for i in new_posts:
        post = i
        post.click()
        path = '//span[@class ="fr66n"]/button'
        wait = WebDriverWait(driver,10)
        wait.until(EC.presence_of_element_located((By.XPATH,path)))
        but1 = driver.find_element_by_xpath('//span[@class ="fr66n"]/button')
        data = BeautifulSoup(but1.get_attribute('outerHTML'),'html.parser')
        if data.svg['aria-label'] == 'Like' :
           but1.click()
        else :
            print('Already liked')
        driver.find_elements_by_class_name('wpO6b ')[-1].click()
    driver.back()
like_posts()
driver.get('https://www.instagram.com')
```

```python
#### 5. 2) Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.
def dislike_posts() :
    open_profile('dilsefoodie')
    time.sleep(2)
    driver.execute_script('window.scrollTo(0,arguments[0]);',0)
    current_height = driver.execute_script('return document.body.scrollHeight;')
    while True :
        driver.execute_script('window.scrollTo(0,arguments[0]);',current_height)
        time.sleep(3)
        new_height = driver.execute_script('return document.body.scrollHeight;')
        new_posts = driver.find_elements_by_class_name('_9AhH0')
        #print(len(new_posts))
        if(len(new_posts)>=30) :
            break;
        if(new_height == current_height) :
            break;
        current_height = new_height
    new_posts = driver.find_elements_by_class_name('_9AhH0')
    driver.execute_script('window.scrollTo(0,arguments[0]);',0)
    new_posts = new_posts[:30]
    #print(new_posts)
    for i in new_posts:
        post = i
        post.click()
        path = '//span[@class ="fr66n"]/button'
        wait = WebDriverWait(driver,10)
        wait.until(EC.presence_of_element_located((By.XPATH,path)))
        but1 = driver.find_element_by_xpath('//span[@class ="fr66n"]/button')
        data = BeautifulSoup(but1.get_attribute('outerHTML'),'html.parser')
        if data.svg['aria-label'] == 'Unlike' :
           but1.click()
        else :
            print('Already disliked')
        driver.find_elements_by_class_name('wpO6b ')[-1].click()
    driver.back()
dislike_posts()
driver.get('https://www.instagram.com')
```

### 6 Extract list of followers
1 Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.

2 Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

```python
#### 6. 1) Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.
from selenium.webdriver.common.keys import Keys  # import KEYS 
def  get_followers(s) :
    open_profile(s) 
    wait = WebDriverWait(driver,10)
    wait.until(EC.presence_of_element_located((By.CLASS_NAME,'-nal3 ')))
    followers =  driver.find_element_by_partial_link_text('followers')
    followers.click()
    while True :
            time.sleep(2)
            new_follow = driver.find_elements_by_class_name('FPmhX')
            if(len(new_follow)>=500) :
                break;
            new_follow[0].send_keys(Keys.END)
    values = []
    for i in new_follow[:500] :
        values.append(i.text)
    return values 
def find_usernames() :
    foodtalk_id = get_followers('foodtalkindia')
    driver.refresh()
    driver.get('https://www.instagram.com/')
    sodelhi_id = get_followers('sodelhi')
    driver.get('https://www.instagram.com/')
    print("***Followers of food ka india***")
    count =0
    for i in foodtalk_id :
        print(count,">",i)
        count += 1
    print("***Followers of so delhi***")
    count = 0
    for i in sodelhi_id :
        print(count,">",i)
        count += 1
    return foodtalk_id
foodfollowers = find_usernames()
```

###  7 Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -
1 If You have already seen the story.

2 Or The user has no story.

3 Or View the story if not yet seen.

```python
def check_story(user_name) :
    open_profile(user_name)
    time.sleep(3)
    try :
        wait = WebDriverWait(driver,10)
        story = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'CfWVH')))
        if int(story.get_attribute('height'))== 209 :
            print('Already seen the story')
        else :
            driver.find_element_by_class_name('h5uC0').click()
            print("Viewing the user story")
    except :
        print(user_name,"has no story at this moment")
check_story('coding.ninjas')
```
